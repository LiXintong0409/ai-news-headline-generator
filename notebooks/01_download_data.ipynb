# 01 - Download THUCNews Dataset

This notebook downloads the THUCNews dataset for Chinese news headline generation.

**Dataset**: THUCNews (Tsinghua University Chinese News)
- 740,000+ news articles across 14 categories
- Source: http://thuctc.thunlp.org/


```python
import os
import urllib.request
import zipfile
from pathlib import Path

# Create data directory
data_dir = Path('../data/raw')
data_dir.mkdir(parents=True, exist_ok=True)

print(f"Data will be saved to: {data_dir.absolute()}")
```

```python
# THUCNews download URL (mirror)
url = "https://github.com/brightmart/nlp_chinese_corpus/raw/master/data/thucnews.zip"
zip_path = data_dir / "thucnews.zip"

print("Downloading THUCNews dataset...")
print("This may take a few minutes...")

urllib.request.urlretrieve(url, zip_path)
print(f"Downloaded to: {zip_path}")
print(f"File size: {zip_path.stat().st_size / 1024 / 1024:.1f} MB")
```

```python
# Extract the zip file
print("Extracting...")
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(data_dir)
print("Extraction complete!")

# List extracted contents
for item in data_dir.iterdir():
    print(f"  {item.name}")
```

```python
# Clean up zip file
zip_path.unlink()
print("Removed zip file, kept extracted data.")
```
